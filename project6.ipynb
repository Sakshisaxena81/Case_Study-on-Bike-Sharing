```python
def rex(x):
    return x.map({1:"jan",2:"feb",3:"mar",4:"apr",5:"may",6:"jun",7:"jly",8:"aug",9:"sep",10:"oct",11:"nov",12:"dec"})
```

```python
df[["mnth"]]=df[["mnth"]].apply(rex)
```

```python
df["mnth"].value_counts()
```

```python
def st(x):
    return x.map({1:"wed",2:"thur",3:"fri",4:"sat",5:"sun",6:"mon",0:"tues"})
```

```python
df[["weekday"]]=df[["weekday"]].apply(st)
```

```python
df["weekday"].value_counts()
```
